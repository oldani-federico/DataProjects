```python
import pandas as pd
pd.set_option("display.max_rows", None, "display.max_columns", None)
import datetime

x = datetime.datetime.now()

Y = str(x.year)[2:]
M = str(x.month)
D = str(x.day)

path = str('C:/Users/fede2/OneDrive - University of Miami/1M/1Money_' + M + '_' + D + '_' + Y + '.csv')
```

```python
data = pd.read_csv(path)
display(data)
```

```python
data = data[:-9]
```

```python
data = data.rename(columns={"DATE":"Date",
                     "TYPE":"Type",
                     "FROM ACCOUNT":"Account",
                     "TO ACCOUNT / TO CATEGORY":"Category",
                     "AMOUNT":"Amount",
                    "NOTES":"Notes"})
```

```python
data
```

```python
data["Subcategory"] = data["Category"]
```

```python
import re

def CleanCategory(c):  
    if re.search('\(.*', c): 
        pos = re.search('\(.*', c).start() 
        return c[pos+1:-1] 
    else: 
        return c
    
data['Subcategory'] = data['Subcategory'].apply(CleanCategory) 
```

```python
data
```

```python
def CleanSubcategory(c):  
    if re.search('\(.*', c): 
        pos = re.search('\(.*', c).start() 
        return c[:pos] 
    else: 
        return c

data['Category'] = data['Category'].apply(CleanSubcategory)
```

```python
data
```

```python
data = data[["Date", "Type", "Account", "Category", "Subcategory", "Amount", "Notes"]]
data
```
